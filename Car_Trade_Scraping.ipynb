```python
import pandas as pd
import numpy as np
import selenium
from selenium import webdriver
import time
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
driver= webdriver.Chrome(r"E:\chromedriver.exe")
time.sleep(4)
```

```python
url = "https://www.cartrade.com/"
driver.get(url)
time.sleep(6)

# creating empty lists for scraping data
name=[]
price=[]
number_of_kms=[]
fuel=[]
location=[]

# We will first scrap the data for Maruti Suzuki

Find_car_section=driver.find_element_by_xpath('//*[@id="rvwtop"]/div/div[1]/div[2]/div[1]/label/span')
Find_car_section.click()

Select_city=driver.find_element_by_xpath('//*[@id="ucity"]/optgroup[1]/option[1]')
Select_city.click()

search_button=driver.find_element_by_xpath('//*[@id="rvwtop"]/div/div[1]/div[2]/div[2]')
search_button.click()

Maruti_suzuki=driver.find_element_by_xpath('//*[@id="Maruti Suzuki"]')
Maruti_suzuki.click()

time.sleep(1)
#scrapping the required details
start=0
end=100
for page in range(start,end):#for loop for scrapping 3 page
    names=driver.find_elements_by_xpath("//h2[@class='h2heading truncate']")
    for i in names:
        name.append(i.text)
        
    prices=driver.find_elements_by_xpath("//div[@class='cr_prc']")
    for i in prices:
        price.append(i.text)
            
    number_of_km=driver.find_elements_by_xpath("//div[@class='info_cr_new']/ul[1]/li[1]")
    for i in number_of_km:
        number_of_kms.append(i.text)
        
    fuels=driver.find_elements_by_xpath("//div[@class='info_cr_new']/ul[1]/li[3]")
    for i in fuels:
        fuel.append(i.text)
        
    locations=driver.find_elements_by_xpath("//div[@class='info_cr_new']/ul[1]/li[5]")
    for i in locations:
        location.append(i.text)
        
#convering all lists to dataframe

Name=pd.DataFrame({'Name':name})
Price=pd.DataFrame({'Price':price})
Number_of_Kms=pd.DataFrame({'Number_of_Kms':number_of_kms})
Fuel=pd.DataFrame({'Fuel':fuel})
Location=pd.DataFrame({'Location':location})

# concatinating all the dataframes

Maruti_Suzuki_Delhi=pd.concat([Name,Price,Number_of_Kms,Fuel,Location],axis=1)

# Removing unwanted words/ signs from dataframe
Maruti_Suzuki_Delhi['Price'] = Maruti_Suzuki_Delhi['Price'].str.replace("Make Offer","")
Maruti_Suzuki_Delhi['Price'] = Maruti_Suzuki_Delhi['Price'].str.replace("₹","")
Maruti_Suzuki_Delhi['Number_of_Kms'] = Maruti_Suzuki_Delhi['Number_of_Kms'].str.replace("KMs","")
Maruti_Suzuki_Delhi
```

```python
#Manufacturing year is present in Name column, we will extract that 
Maruti_Suzuki_Delhi['Manufacture_year'] = Maruti_Suzuki_Delhi['Name'].str[:4]

# We also know that all the cars are Maruti Suzuki make, hence we will create a new column under the name of Make
Maruti_Suzuki_Delhi["Make"]=np.nan
Maruti_Suzuki_Delhi["Make"]=Maruti_Suzuki_Delhi["Make"].fillna("Maruti Suzuki")
Maruti_Suzuki_Delhi
```
